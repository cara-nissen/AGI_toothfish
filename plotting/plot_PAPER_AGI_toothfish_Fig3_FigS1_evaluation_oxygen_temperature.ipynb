```python
#---
# load WOA data
#---

path_woa = '/pscratch/sd/c/cnissen/WOA18/'

f1 = Dataset(path_woa+'woa18_all_o00_01.nc') # already in mumol kg-1  !!!!!!
oxy_woa   = np.squeeze(f1.variables['o_an'][:,:,:,:]) # depth x lat x lon
lat_woa   = np.squeeze(f1.variables['lat'][:])
lon_woa   = np.squeeze(f1.variables['lon'][:])
depth_woa = np.squeeze(f1.variables['depth'][:])
f1.close()

f1 = Dataset(path_woa+'woa18_decav_t00_01.nc') # in-situ temperature!
temp_woa  = np.squeeze(f1.variables['t_an'][:,:,:,:]) # depth x lat x lon
f1.close()

f1 = Dataset(path_woa+'woa18_decav_s00_01.nc') # practical salinity
salt_woa  = np.squeeze(f1.variables['s_an'][:,:,:,:]) # depth x lat x lon
f1.close()

oxy_woa = oxy_woa/1e6 # -> this is needed to get pO2 in the right order of magnitude
print('oxy_woa.shape:',oxy_woa.shape,np.min(oxy_woa),np.max(oxy_woa))
print('temp_woa.shape:',temp_woa.shape)
print('salt_woa.shape:',salt_woa.shape)
print(depth_woa.shape,lat_woa.shape,lon_woa.shape)

#----
# get pO2 from in-situ temp and oxygen in mumol kg-1
#----            
    
#-----
# constants
#-----
xO2 = 0.20946 # O2 mole fraction dry air
Vm = 0.317 # partial molar volume O2 (m3 mol-1 Pa dbar-1)
R = 8.3145 # gas constant in m3 Pa K-1 mol -1
KC = 273.15 # deg C to Kelvin

A0 = 5.80871
A1 = 3.20291
A2 = 4.17887
A3 = 5.10006
A4 = -0.0986643
A5 = 3.80369
B0 = -0.00701577
B1 = -0.00770028
B2 = -0.0113864
B3 = -0.00951519
C0 = -0.000000275915
D0 = 24.4543
D1 = -67.4509
D2 = -4.8489
D3 = -5.44*1e-4

lat_3D    = np.tile(lat_woa,[depth_woa.shape[0],lon_woa.shape[0],1]) # (99, 91, 720) depth, lat, lon
depths_3D = np.tile(depth_woa,[lon_woa.shape[0],lat_woa.shape[0],1])
print(lat_3D.shape,depths_3D.shape)
depths_3D = np.transpose(depths_3D,[2,0,1]) # (99, 91, 720) depth, lat, lon

pres_3D   = sw.eos80.pres(depths_3D,lat_3D) # (99, 91, 720) depth, lat, lon
print(pres_3D.shape)
pres_3D = np.transpose(pres_3D,[0,2,1])

ref_pressure = 0
#data_tt_woa = sw.eos80.temp(salt_woa,temp_woa,pres_3D,ref_pressure)
#print(data_tt_woa.shape)

## scaled temp
T_scaled = np.log((298.15-temp_woa)/(KC+temp_woa)) # np.log -> ln(), np.log10 -> log10

# saturation concentration of O2 in seawater
A = A0 + A1*T_scaled + A2*T_scaled*T_scaled + A3*T_scaled*T_scaled*T_scaled+\
                    A4*T_scaled*T_scaled*T_scaled*T_scaled +\
                    A5*T_scaled*T_scaled*T_scaled*T_scaled*T_scaled +\
                    salt_woa*(B0+B1*T_scaled+B2*T_scaled*T_scaled+B3*T_scaled*T_scaled*T_scaled)+\
                    C0*salt_woa*salt_woa
c_sat_o2 = 1e-6*np.exp(A)
#
# water vapour pressure
pH2O = 1013.25 * np.exp(D0 + D1*(100./(temp_woa+KC))+D2*np.log((temp_woa+KC)/100.)+D3*salt_woa)

K0 = c_sat_o2/(xO2 * (1013.25 - pH2O))
pO2_woa = (oxy_woa/K0)*np.exp((Vm*pres_3D)/(R*(KC+temp_woa)))
pO2_woa = np.ma.masked_where(temp_woa.mask==True,pO2_woa)
print('Min/Max pO2 in WOA:',np.min(pO2_woa),np.max(pO2_woa))

print('done')
```

```python
#-----
# FOR MAPS, load fields on original model grid
# load data for 1995-2014: oxygen
#-----
# also load rho for each year (in-situ density) to convert oxygen from mmol m-3 to mumol kg-1
# calculate pO2

vari  = 'bgc22'
vari2 = 'rho'
vari3  = 'thetao'
vari4 = 'so'

path  = '/pscratch/sd/c/cnissen/COARZE_oxy/hist/'
path2 = '/pscratch/sd/c/cnissen/COARZE_rho/hist/'
path3  = '/pscratch/sd/c/cnissen/COARZE_temp/hist/'
path4 = '/pscratch/sd/c/cnissen/COARZE_salt/hist/'

years = np.arange(1995,2014+1,1)

array_1990s_oxy   = np.zeros([len(np.unique(zlevs_model)),len(lat_model),len(years)]) 
array_1990s_pO2   = np.zeros([len(np.unique(zlevs_model)),len(lat_model),len(years)]) 
for yy in range(0,len(years)):
    print ('Load '+str(years[yy])+' of '+which_sim)
    
    # oxygen
    f1 = Dataset(path+vari+'_fesom_'+str(years[yy])+'0101.nc')
    data1 = f1.variables[vari][:,:,:].mean(axis=0) # depth x location, calculate annual mean here
    f1.close()
    data1 = np.ma.masked_where(data1<-999,data1)
    data1[data1.mask==True]=0
    # rho
    f1 = Dataset(path2+vari2+'_fesom_'+str(years[yy])+'0101.nc')
    data2 = f1.variables[vari2][:,:,:].mean(axis=0) # depth x location, calculate annual mean here
    f1.close()
    data2 = np.ma.masked_where(data2<-999,data2)
    data2[data2.mask==True]=0   
    # temp
    f1 = Dataset(path3+vari3+'_fesom_'+str(years[yy])+'0101.nc')
    data3 = f1.variables[vari3][:,:,:].mean(axis=0) # depth x location, calculate annual mean here
    f1.close()
    data3 = np.ma.masked_where(data3<-999,data3)
    data3[data3.mask==True]=0
    # salt
    f1 = Dataset(path4+vari4+'_fesom_'+str(years[yy])+'0101.nc')
    data4 = f1.variables[vari4][:,:,:].mean(axis=0) # depth x location, calculate annual mean here
    f1.close()
    data4 = np.ma.masked_where(data4<-999,data4)
    data4[data4.mask==True]=0
    insitu = sw.eos80.temp(data4,data3,pres_3D_FESOM,ref_pressure)
    
    array_1990s_oxy[:,:,yy]    = 1000*data1/data2
    
    #---
    # get pO2
    #---
    ## scaled temp
    T_scaled = np.log((298.15-insitu)/(KC+insitu)) # np.log -> ln(), np.log10 -> log10

    # saturation concentration of O2 in seawater
    A = A0 + A1*T_scaled + A2*T_scaled*T_scaled + A3*T_scaled*T_scaled*T_scaled+\
                        A4*T_scaled*T_scaled*T_scaled*T_scaled +\
                        A5*T_scaled*T_scaled*T_scaled*T_scaled*T_scaled +\
                        data4*(B0+B1*T_scaled+B2*T_scaled*T_scaled+B3*T_scaled*T_scaled*T_scaled)+\
                        C0*data4*data4
    c_sat_o2 = 1e-6*np.exp(A)
    #
    # water vapour pressure
    pH2O = 1013.25 * np.exp(D0 + D1*(100./(insitu+KC))+D2*np.log((insitu+KC)/100.)+D3*data4)

    K0 = c_sat_o2/(xO2 * (1013.25 - pH2O))
    array_1990s_pO2[:,:,yy] = (1e-6*array_1990s_oxy[:,:,yy]/K0)*np.exp((Vm*pres_3D_FESOM)/(R*(KC+insitu)))
    array_1990s_pO2[:,:,yy] = np.ma.masked_where(insitu.mask==True,array_1990s_pO2[:,:,yy])
    print('Min/Max pO2 in FESOM:',np.min(array_1990s_pO2[:,:,yy]),np.max(array_1990s_pO2[:,:,yy]))
    
    del data1,data2
    
# decadal avg
array_1990s_oxy    = np.nanmean(array_1990s_oxy,axis=2) 
array_1990s_pO2    = np.nanmean(array_1990s_pO2,axis=2) 

# correct pO2 at locations that are not available
array_1990s_pO2[array_1990s_oxy==0] = 0

print ('done')
```
